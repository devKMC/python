### seaborn을 활용한 1차원 분포 플롯

seaborn 패키지에서 제공해주는 여러 가지 메서드를 활용하여 간단하게 1차원 분포  
플롯을 표시 할 수 있음

### 1차원 실수 분포 플롯

seaborn 패키지는 히스토그램 형식과는 다른 커널 밀도, 러그 플롯을 제공함
`rugplot()` : 데이터 위치를 x축 위에 작은 선분으로 표시( 러그 플롯 )  
`kdeplot()` : 히스토그램과 비슷하지만 곡선형태로 표시 ( 커널 플롯 )  

```python
data = iris.petal_length.values
data
```

```python
seaborn.rugplot(data)
plt.title('러그 플롯')
plt.grid(True)
plt.show()
```

```python
seaborn.kdeplot(data)
plt.title('커널 플롯')
plt.grid(True)
plt.show()
```

### seaborn을 활용한 다차원 데이터

다차원 데이터는 3가지의 경우의 수를 가질 수 있음

- 모든 데이터가 실수형
- 모든 데이터가 카테고리형
- 데이터가 실수형과 카테고리형이 섞여 있음

#### 2차원 실수형 데이터

데이터가 2차원이면서 모두 실수이면 `jointplot()` 메서드를 사용하여 표현할 수 있음  
`jointplot()`: 스캐터 플롯, 등고선 플롯 등을 나타낼 수 있고,  
양측에 히스토그램을 표현해 줌 (데이터프레임만 사용 가능)

```python
iris
```

```python
seaborn.jointplot(data=iris, x='sepal_length', y= 'sepal_width', kind = 'scatter')
plt.suptitle('꽃 받침의 길이와 넓이 비교 joint Plot', y=1.02)
plt.grid(True)
plt.show()
```

```python
seaborn.jointplot(data=iris, x='sepal_length', y= 'sepal_width', kind ='scatter')
plt.suptitle('꽃 받침의 길이와 넓이 비교 joint Plot', y=1.02)
plt.grid(True)
plt.show()
```

#### 다차원 실수형 데이터  

3차원 이상의 데이터라면 `pairplot()` 메서드로 그리드 형태의 스캐터 플롯의 행렬을 만들어 주지만,  
같은 데이터가 만나는 대각선에는 각 데이터의 히스토그램을 표현

```python
seaborn.pairplot(iris)
plt.title('다차원 pair plot', y=1.02)
plt.show()
```

카테고리형 데이터가 포함되어 있는 경우라면  
`hue` 인수에 카테고리 변수 이름을 지정함

```python
seaborn.pairplot(iris, hue='species', markers=['o','s','D'])
plt.suptitle('다차원 pair plot', y=1.02)
plt.show()
```

#### 2차원 카테고리 데이터 

데이터가 2차원이면서 모두 카테고리 값이면 `heatmap()` 메서드로 히트맵을 표현 할 수 있음  

```python
titanic_size = titanic.pivot_table(
    index = 'class', columns= 'sex', aggfunc='size'

)
titanic_size
```

```python
seaborn.heatmap(titanic_size, cmap=seaborn.light_palette('gray',as_cmap=True)
                , annot= True, fmt='d')
plt.title('히트맵')
plt.show()
```

#### 2차원 복합 데이터
데이터가 2차원이면서 실수 값과 카테고리 값이 섞여 있다면  
기존 플롯 외의 다른 플롯을 사용할 수 있음

`barplot()` : 카테고리 값에 따른 실수 값의 평균과 표준 편차를 바 형태로 표시  
평균은 막대 그래프로, 편차는 에러바로 표현  

```python
seaborn.barplot(data = tips, x= 'day', y='total_bill') 
plt.title('바 플롯')
plt.grid(True)
plt.show()
```

`boxplot()` 사분위수와 아웃라이어라는 데이터를 표현하는 방법

```python
seaborn.boxplot(data = tips , x= 'day', y= 'total_bill')
plt.title('박스 플롯')
plt.grid(True)
plt.show()
```

`violinplot()` : 세로 방향으로 커널 밀도 히스토그램을 그려줌

```python
seaborn.violinplot(data=tips, x= 'day',y='total_bill')
plt.title('바이올린 플롯')
plt.grid(True)
plt.show()
```

`stripplot()` : 모든 데이터를 점으로 표시해서 보여줌

```python
seaborn.stripplot(data=tips, x= 'day',y='total_bill')
plt.title('스트립 플롯')
plt.grid(True)
plt.show()
```

`swarmplot()` : 스트립 플롯에서 데이터가 겹치는 부분을 겹치지 않게 표현 

```python
seaborn.swarmplot(data=tips, x='day',y='total_bill')
plt.title('스웜 플롯')
plt.grid(True)
plt.show()
```

### seaborn을 활용한 다차원 복합 데이터 

seaborn에서 제공해주는 플롯 메서드들은 `hue` 인수로 카테고리 값을 추가로 지정 할 수 있음  

```python
seaborn.violinplot(data = tips, y= 'total_bill', hue = 'sex') 
plt.title('바 플롯')
plt.grid(True)
plt.show()
```

`stripplot()`, `swarmplot()`, `violinplot()` 은 `split`, `dodge` 인수로  
시각화 방법을 변경할 수 있음

```python
seaborn.violinplot(data = tips, y= 'total_bill',hue='sex',split = True) 
plt.title('바 플롯')
plt.grid(True)
plt.show()
```

```python
seaborn.violinplot(data = tips, y= 'total_bill', hue = 'sex',dodge=True) 
plt.title('바 플롯')
plt.grid(True)
plt.show()
```

`heatmap()` 으로도 두 개의 케토길 값에 의한 실수 값 변화를 볼 수 있음

```python
flights_passengers = flights.pivot(index="month", columns="year", values="passengers")
plt.title("연도, 월 별 승객수에 대한 Heatmap")
seaborn.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```

### seaborn을 활용한 catplot

색상과 행을 사용해서 3개 이상의 카테고리 값의 분포를 표현  


```python
data = titanic[titanic.survived.notnull()]
seaborn.catplot(data=data, x="age", y="sex", hue="survived", row="class",   # hue = 색상 구분
            kind="violin", split=True, height=2, aspect=4)
plt.suptitle("캣 플롯" ,y= 1.0
             )
plt.grid(True)
plt.show()
```

### seaborn을 활용한 차트 겹치기
시각적 효과를 높이기 위해 여러 종류의 차트를 겹쳐서 표기 가능

```python
plt.title('박스 플롯 + 스트립 플롯')
seaborn.boxplot(data=tips, x= 'tip', y='day' ,whis=np.inf)  #  whis 를 통해 벗어나는 범위는 안보이게 만듬
seaborn.stripplot(data =tips, x= 'tip', y= 'day')
plt.grid(True)
plt.show()
```

```python
plt.title("Violin plot과 Swarm Plot로 표현한, 요일 별 팁")
seaborn.violinplot(x="day", y="total_bill", data=tips, inner=None)
seaborn.swarmplot(x="day", y="total_bill", data=tips, color="0.9")
plt.grid(True)
plt.show()
```

### seaborn을 활용한 스타일 변경

seaborn 패키지는 matplotlib에서 제공하는 기본 스타일이 아닌 특정 스타일로 변경할 수 있도록 함  
`set()` 메서드로 색상, 틱 스타일 등을 변경할 수 있음  
`set_style()` 메서드로 틱 스타일 변경할 수 있음

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for number in range (1,7):
        plt.plot (x, np.sin(x + number * 0.5) * (7 - number) * flip)

plt.show()
```

```python
seaborn.set_style('ticks')
sinplot()
```

```python
seaborn.set_style('darkgrid')
sinplot()
```

```python
seaborn.set_style('dark')
sinplot()
```
