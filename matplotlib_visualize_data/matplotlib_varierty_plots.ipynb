### 바 차트 (막대 그래프)
x 축의 값이 분류가 가능한 카테고리 형태일 때 `bar()` , `barh()` 메서드로 막대 그래프를 표시할 수 있음  

주의 : bar 메서드는 x축의 값이 첫 번째 인수로 옴  

```python
# matplotlib 한글 표시 설정
import matplotlib.font_manager as fm

# matplotlib의 글꼴 관리자에 특정 글꼴 추가
fe=fm.FontEntry(
    fname=r'font/NanumGothic.ttf',          # ttf 파일이 저장되어 있는 경로
    name='NanumGothic')                       
fm.fontManager.ttflist.insert(0, fe)              # Matplotlib에 폰트 추가

#pyplot에 matplotlib의 글꼴 관리자에 등록된 글꼴 적용
plt.rcParams.update({'font.size': 14, 'font.family': 'NanumGothic'})

#pyplot에서 `-`를 문자로 취급하도록 설정
plt.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import numpy as np
```

```python
y = [2,3,1]
x = [1,2,3]
xlabel = ['가','나','다']

plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.show()
```

`xerr` , `yerr` 인수를 지정하면 에러 바를 추가할 수 있음  
`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음 (0: 완전 투명, 1: 완전 불투명)  

```python
np.random.seed(0)
people = ['몽룡','춘향','방자','향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar h')              # 타이틀을 정하고 그래프 생성하는 명령어
plt.barh(y_position, y_performance, xerr=error,alpha=0.4)
plt.yticks(y_position,people)
plt.xlabel('x 축')
plt.grid(True)
plt.show()
```

### 스템 플롯
이산확률이나 자기 상관관계를 표현하고자 할 때 사용하는   
`stem()` 메서드로 표시할 수 있음

```python
x = np.linspace(0.1,2 * np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x), '-.')
plt.grid()
plt.show()

x
```

### 파이 차트
각 분류가 전체에서 얼만큼 차지하고 있는지 파악하려 할때 사용하는 차트   
`pie()` 메서드로 표현할 수 있음  

주의 : 차트가 원 형태를 유지할 수 있게 하려면 `axis('equal')을 지정해야 함

```python
label = ['개구리','돼지','개','고영이']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','#FF0090']
explode = (0,0.1,0,0)
plt.title('파이 차트')
plt.pie(sizes,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

### 히스토그램
특정 구간을 정하여 그 구간에 대한 집계 데이터를 표시할 때 사용하는 차트  
`hist()` 메서드로 표시할 수 있음  
`bins()` 인수로 집계할 구간의 정보를 받음  

np.random.randn(): 이 함수는 평균이 0이고 표준 편차가 1인 표준 정규 분포(가우시안 분포)에서 난수를 생성합니다.  
따라서 반환되는 난수의 범위는 보통 -3에서 3까지이며, 이것은 99.7%의 확률로 발생합니다.

np.random.rand(): 이 함수는 0과 1 사이의 균일 분포에서 난수를 생성합니다. 즉, `(0, 1)` 범위의 실수를 균일하게 생성합니다.

```python
x = np.random.randn(1000)

plt.title('히스토그램')
arrays, bins, patches = plt.hist(x, bins=10)
plt.grid(True)
plt.show()

arrays
```

### 스케터 플롯
두 변수 간의 상관관계를 나타내고자 할 때 사용하는 차트  
`scatter()` 메서드로 표현할 수 있음  

점 하나의 위치는 x,y 값으로 표현됨

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('스캐터 플롯')
plt.scatter(x,y)
plt.grid(True)
plt.show()
```

만약 x에 대해서 여러 변수를 한 차트로 비교하고자 할 때 점의 크기 혹은 색상 추가로 표현할 수 있음  
'scatter()' 메서드의 `c`인수로 색상을, `s` 인수로 크기를 지정할 수 있음  

```python
n = 30
x = np.random.randn(n)
y1 = np.random.randn(n)
y2 = np.random.randn(n)
y3 = np.pi * (15 * np.random.rand(n))**2

plt.title('버블 차트')
plt.scatter(x,y1,c=y2,s=y3)
plt.grid(True)
plt.show()
```

### imshow  

이미지 데이터 처럼 행과 열을 가진 2차원 데이터에 대하여 각 행렬 값을 색으로 표현하는 차트  
`imshow()` 메서드로 표현할 수 있음

```python
# pip install scikit_learn 실행이 안될시에 먼저 인스톨후 사용해야함
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python

x = digits.images[0]   # 숫자 변경시 그림 변경
x

plt.title('mnist digits.0')
plt.imshow(x, interpolation = 'nearest', cmap =plt.cm.bone_r)
plt.subplots_adjust(left=0.35,right=0.65, bottom = 0.35 , top= 0.65)            # 크기(사이즈) 조절
plt.xticks([])
plt.yticks([])
plt.show()
```

`cmap` 인수로 데이터를 표현하는 칼라맵을 지정할 수 있음  

`inter polation` 인수로 데이터의 시각화 표현을 지정할 수 있음

```python
figure, axes = plt.subplots(1, 4, figsize=(12, 3), subplot_kw={'xticks': [], 'yticks': []})

axes[0].set_title('color map Blues')
axes[0].imshow(x, interpolation='nearest', cmap=plt.cm.Blues)
axes[1].set_title('color map Blues_r')
axes[1].imshow(x, interpolation='nearest', cmap=plt.cm.Blues_r)

axes[2].set_title('interpolation none')
axes[2].imshow(x, interpolation='none', cmap=plt.cm.bone_r)
axes[3].set_title('interpolation kaiser')
axes[3].imshow(x, interpolation='kaiser', cmap=plt.cm.bone_r)

plt.show()
```

```python
def f(x,y):
    return (1-x / 2 + x** 5 + y ** 3) * np.exp(-x ** 2-y ** 2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
xx, yy = np.meshgrid(x,y)
zz = f(xx,yy)


plt.title('컨투어 플롯')
# plt.contour(xx,yy,zz, colors = 'black')
plt.contourf(xx,yy,zz,alpha=0.75, cmap='jet')
plt.grid(True)

plt.show()
```

### 3D 서피스 플롯
3차원 데이터를 입체적으로 표현하는 차트  
3d axes 객체를 생성하여 사용함  

`plot_wireframe()', 'plot_surface()` 메서드로 표현

```python
x = np.arange(-4,4,0.25)
x = np.arange(-4,4,0.25)
xx, yy = np.meshgrid(x,y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

axes = plt.axes(projection='3d')
axes.set_title('3D 서피드 플롯')
axes.plot_surface(xx,yy,zz,rstride=1,cstride=1 , cmap= 'hot')
plt.show()
```
