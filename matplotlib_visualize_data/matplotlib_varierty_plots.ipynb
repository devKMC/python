### 바 차트 (막대 그래프)
x 축의 값이 분류가 가능한 카테고리 형태일 때 `bar()` , `barh()` 메서드로 막대 그래프를 표시할 수 있음  

주의 : bar 메서드는 x축의 값이 첫 번째 인수로 옴  

````python
# matplotlib 한글 표시 설정
import matplotlib.font_manager as fm

# matplotlib의 글꼴 관리자에 특정 글꼴 추가
fe=fm.FontEntry(
    fname=r'font/NanumGothic.ttf',          # ttf 파일이 저장되어 있는 경로
    name='NanumGothic')                       
fm.fontManager.ttflist.insert(0, fe)              # Matplotlib에 폰트 추가

#pyplot에 matplotlib의 글꼴 관리자에 등록된 글꼴 적용
plt.rcParams.update({'font.size': 14, 'font.family': 'NanumGothic'})

#pyplot에서 `-`를 문자로 취급하도록 설정
plt.rcParams['axes.unicode_minus'] = False
````

````python
import pandas as pd
import numpy as np
````

````python
y = [2,3,1]
x = [1,2,3]
xlabel = ['가','나','다']

plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.show()
````

`xerr` , `yerr` 인수를 지정하면 에러 바를 추가할 수 있음  
`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음 (0: 완전 투명, 1: 완전 불투명)  

````python
np.random.seed(0)
people = ['몽룡','춘향','방자','향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar h')              # 타이틀을 정하고 그래프 생성하는 명령어
plt.barh(y_position, y_performance, xerr=error,alpha=0.4)
plt.yticks(y_position,people)
plt.xlabel('x 축')
plt.grid(True)
plt.show()
````

### 파이 차트
각 분류가 전체에서 얼만큼 차지하고 있는지 파악하려 할때 사용하는 차트   
`pie()` 메서드로 표현할 수 있음  

주의 : 차트가 원 형태를 유지할 수 있게 하려면 `axis('equal')을 지정해야 함

````python
label = ['개구리','돼지','개','고영이']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','#FF0090']
explode = (0,0.1,0,0)
plt.title('파이 차트')
plt.pie(sizes,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
````

### 히스토그램
특정 구간을 정하여 그 구간에 대한 집계 데이터를 표시할 때 사용하는 차트  
`hist()` 메서드로 표시할 수 있음  
`bins()` 인수로 집계할 구간의 정보를 받음  

np.random.randn(): 이 함수는 평균이 0이고 표준 편차가 1인 표준 정규 분포(가우시안 분포)에서 난수를 생성합니다.  
따라서 반환되는 난수의 범위는 보통 -3에서 3까지이며, 이것은 99.7%의 확률로 발생합니다.

np.random.rand(): 이 함수는 0과 1 사이의 균일 분포에서 난수를 생성합니다. 즉, `(0, 1)` 범위의 실수를 균일하게 생성합니다.

````python
x = np.random.randn(1000)

plt.title('히스토그램')
arrays, bins, patches = plt.hist(x, bins=10)
plt.grid(True)
plt.show()

arrays
````

### 스케터 플롯
두 변수 간의 상관관계를 나타내고자 할 때 사용하는 차트  
`scatter()` 메서드로 표현할 수 있음  

점 하나의 위치는 x,y 값으로 표현됨

````python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('스캐터 플롯')
plt.scatter(x,y)
plt.grid(True)
plt.show()
````

만약 x에 대해서 여러 변수를 한 차트로 비교하고자 할 때 점의 크기 혹은 색상 추가로 표현할 수 있음  
'scatter()' 메서드의 `c`인수로 색상을, `s` 인수로 크기를 지정할 수 있음  

````python
n = 30
x = np.random.randn(n)
y1 = np.random.randn(n)
y2 = np.random.randn(n)
y3 = np.pi * (15 * np.random.rand(n))**2

plt.title('버블 차트')
plt.scatter(x,y1,c=y2,s=y3)
plt.grid(True)
plt.show()
````

### 3D 서피스 플롯
3차원 데이터를 입체적으로 표현하는 차트  
3d axes 객체를 생성하여 사용함

````python
x = np.arange(-4,4,0.25)
x = np.arange(-4,4,0.25)
xx, yy = np.meshgrid(x,y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

axes = plt.axes(projection='3d')
axes.set_title('3D 서피드 플롯')
axes.plot_surface(xx,yy,zz,rstride=1,cstride=1 , cmap= 'hot')
plt.show()
````

### 삼각 그리드 생성
`Triangulation` 클래스로 삼각 그리드를 생성할 수 있음
`Triangulation` 클래스의 생성자는 x,y,triangles 세 개의 인수를 받음
`triangles`는 각 점의 기하학적 위상 정보

````python
x = np.array([0,1,2])
y = np.array([0, np.sqrt(3),0])
triangles = [[0,1,2]]
triangle = mtri.Triangulation(x,y,triangles)

plt.title('삼각 그리드')
plt.triplot(triangle, 'ko-')
plt.xlim(-0.1,2.1)
plt.ylim(-0.1,1.8)
plt.grid(True)
plt.show
````

````python

x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2*np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]   # 선긋는 번호 지정
triang = mtri.Triangulation(x, y, triangles)            # 주어진 점의 집합을 사용하여 삼각형 메쉬를 만드는 데 사용됩니다.
plt.title("여러개의 삼각형 그리기")
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)
````

### 삼각 그리드 세분화
`TriRefiner`, `UniformTriRefiner` 클래스로 그리드를 세분화 할 수 있음
`refiner_triangulation()` : 단순히 삼각 그리드를 세분화
`refine_field()` : 실제 함수에 최적화된 삼각 그리드 생성

````python
refiner = mtri.UniformTriRefiner(triang)
triangle2 = refiner.refine_triangulation(subdiv=1)


plt.title("삼각 그리드 세분화")
plt.triplot(triangle2, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)
plt.show()
````

### 삼각 그리드 플롯
삼차원 그리드 상에 `tricontour` , `tricontourf` , `plot_trisur`, `plot_trisurf`를 사용하여  
2차원 등고선이나 3차원 표면을 그릴 수 있음  

````python
triangle3 = refiner.refine_triangulation(subdiv=5)
z = np.cos(1.5 * triangle3.x) * np.cos(1.5 * triangle3.y)

plt.title('삼각 그리드 등고선')
plt.tricontourf(triangle3,z,cmap='jet')
plt.grid(True)
plt.show()
````

````python
triangle4 = refiner.refine_triangulation(subdiv=3)
z = np.cos(1.5 * triangle4.x) * np.cos(1.5 * triangle4.y)

axes = plt.axes(projection="3d")
axes.set_title("삼각 그리드 3차원")
axes.plot_trisurf(triangle4.x, triangle4.y, z, cmap='jet', linewidth=0.2)
axes.tricontourf(triangle4, z, zdir='z', offset=-1.2, cmap='coolwarm')
axes.set_zlim(-1, 1)
axes.view_init(40, -40)
````
